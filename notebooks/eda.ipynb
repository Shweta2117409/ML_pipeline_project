# Heart Disease — Exploratory Data Analysis

This notebook explores the raw dataset before model training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
df = pd.read_csv('../data/raw.csv')
print(df.shape)
df.head()
```

```python
# Basic info
df.info()
df.describe()
```

```python
# Missing values heatmap
plt.figure(figsize=(12, 2))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='Reds')
plt.title('Missing Values Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Target distribution
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df['target'].value_counts().plot(kind='bar', ax=axes[0], color=['#10b981','#e84c6b'])
axes[0].set_title('Target Distribution (Count)')
df['target'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=['#10b981','#e84c6b'])
axes[1].set_title('Target Distribution (%)')
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 9))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r', center=0, linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Continuous feature distributions by target
cont_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
fig, axes = plt.subplots(1, len(cont_cols), figsize=(18, 4))
for ax, col in zip(axes, cont_cols):
    for t, color in [(0, '#10b981'), (1, '#e84c6b')]:
        df[df['target'] == t][col].dropna().plot(kind='kde', ax=ax, color=color, label=f'Target={t}')
    ax.set_title(col)
    ax.legend(fontsize=8)
plt.suptitle('Feature Distributions by Target', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# IQR Outlier Summary
print('\nOutlier count per continuous column (IQR method):')
for col in cont_cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    n = ((df[col] < q1 - 1.5*iqr) | (df[col] > q3 + 1.5*iqr)).sum()
    print(f'  {col}: {n} outliers')
```
